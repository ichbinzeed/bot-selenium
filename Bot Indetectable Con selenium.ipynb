```python
convertir = lambda selector: '.' + selector.replace(" ", ".")
```

```python
clase = convertir("cy5jw6o atm_5j_223wjw atm_70_87waog atm_j3_1u6x1zy atm_jb_4shrsx atm_mk_h2mmj6 atm_vy_7abht0 dir dir-ltr")
cabania = driver.find_elements(By.CSS_SELECTOR, clase)
print(len(cabania))
```

```python
print(cabania[0].text.split("\n")) #ANALIZO LA ESTRUCTURA
```

```python
datos = [dato.text.split("\n") for dato in cabania]
datos[0]
```

```python
for i in range(len(datos)):
    if datos[i][0] == 'Favorito entre huéspedes':
        datos[i] = datos[i][2:]
    print(datos[i])
```

```python
def obtener_dato(dato):
    try:
        return dato[9]
    except IndexError:
        return None

nombres = [dato[0] for dato in datos]
distancias = [dato[1] for dato in datos]
precios = [dato[5] for dato in datos]
promedios = [obtener_dato(dato) for dato in datos]
```

```python
df = pd.DataFrame({
    'Nombre': nombres,
    'Distancia': distancias,
    'Precio': precios,
    'Promedio': promedios
})

print(df)
```

```python
df.tail(20)
```

```python
df.Precio = df["Precio"].str.replace('USD', '').str.replace('$', '').str.replace(' ', '').astype(int)
df.Precio
```

```python
df.Promedio = df.Promedio.str.replace(",",".").astype(float)
df.Promedio
```
